```python
from binance.client import Client
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

```python
client = Client()
```

```python
data = client.get_klines(symbol='BTCUSDT', interval=Client.KLINE_INTERVAL_1MINUTE)
data[0]
```

```python
time = [datetime.fromtimestamp(point[0] / 1000) for point in data]
```

```python
highPrice = [float(point[1]) for point in data]
```

```python
plt.plot(time, highPrice)
```

```python
df = pd.DataFrame(data=data, columns=["open_time", "open", "high", "low", "close", "volume", "close_time", "quote_volume", "trade_count", "buy_base", "buy_take", "ignore"])
```

```python
df["open"] = pd.to_numeric(df["open"])
df["high"] = pd.to_numeric(df["open"])
df["low"] = pd.to_numeric(df["low"])
df["close"] = pd.to_numeric(df["close"])
df["open_time"] = pd.to_datetime(df["open_time"], unit="ms")
df["close_time"] = pd.to_datetime(df["close_time"], unit="ms")
```
